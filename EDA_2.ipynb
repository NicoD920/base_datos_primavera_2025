### Analizamos los valores faltantes seg√∫n el status

```python
# Identify columns with null values
cols_with_nulls = ['order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date']

# Group by order_status and count nulls in date columns
nulls_by_status = df_orders.groupby('order_status')[cols_with_nulls].apply(lambda x: x.isnull().sum())

display(nulls_by_status)

# Print the total number of records in df_orders
print(f"\nTotal number of records in df_orders: {df_orders.shape[0]}")
```

```python
initial_count = len(df_orders)

# Rule 1
condition_approved_null_and_status = (
    df_orders['order_approved_at'].isnull() &
    ~df_orders['order_status'].isin(['canceled', 'unavailable', 'created'])
)
removed_rule1 = condition_approved_null_and_status.sum()
df_orders = df_orders[~condition_approved_null_and_status].copy()

# Rule 2
condition_carrier_null_and_status = (
    df_orders['order_delivered_carrier_date'].isnull() &
    df_orders['order_status'].isin(['delivered', 'approved'])
)
removed_rule2 = condition_carrier_null_and_status.sum()
df_orders = df_orders[~condition_carrier_null_and_status].copy()

# Rule 3
condition_customer_null_and_status = (
    df_orders['order_delivered_customer_date'].isnull() &
    df_orders['order_status'].isin(['delivered'])
)
removed_rule3 = condition_customer_null_and_status.sum()
df_orders = df_orders[~condition_customer_null_and_status].copy()

final_count = len(df_orders)

print(f"Registros iniciales: {initial_count}")
print(f"Eliminados por Regla 1: {removed_rule1}")
print(f"Eliminados por Regla 2: {removed_rule2}")
print(f"Eliminados por Regla 3: {removed_rule3}")
print(f"Registros finales: {final_count}")
print(f"Porcentaje eliminado total: {(initial_count - final_count)/initial_count:.2%}")

print("\nValores nulos restantes:")
display(df_orders.isnull().sum())
```

```python
# --- 0) Preparaci√≥n: tipos y limpieza -----------------------------------------
for d in [df_orders, df_items, df_reviews, df_payments]:
    d['order_id'] = d['order_id'].astype(str).str.strip()

# (si tienes customers/sellers/products y aparecen en items/orders, al√≠nea tipos)
if 'customer_id' in df_orders.columns:
    df_orders['customer_id'] = df_orders['customer_id'].astype(str).str.strip()
if 'seller_id' in df_items.columns:
    df_items['seller_id'] = df_items['seller_id'].astype(str).str.strip()
if 'product_id' in df_items.columns:
    df_items['product_id'] = df_items['product_id'].astype(str).str.strip()

# --- 1) IDs remanentes y removidos --------------------------------------------
remaining_order_ids = set(df_orders['order_id'].unique())

# Si quieres auditar:
all_order_ids_before = set()
for d in [df_items, df_reviews, df_payments]:
    all_order_ids_before |= set(d['order_id'].unique())
removed_order_ids = all_order_ids_before - remaining_order_ids  # auditor√≠a

# --- 2) Filtrar tablas dependientes por order_id -------------------------------
def filter_by_order_id(df, name):
    initial = len(df)
    df = df[df['order_id'].isin(remaining_order_ids)].copy()
    removed = initial - len(df)
    print(f"Removed {removed} rows from {name} to maintain consistency.")
    return df

df_items    = filter_by_order_id(df_items,    "df_items")
df_reviews  = filter_by_order_id(df_reviews,  "df_reviews")
df_payments = filter_by_order_id(df_payments, "df_payments")

# --- 3) (Opcional pero recomendado) Cascada hacia cat√°logos --------------------
# Si vas a trabajar SOLO con registros efectivamente "vigentes":
# customers que aparecen en orders remanentes
if 'customer_id' in df_orders.columns and 'df_customers' in globals():
    kept_customer_ids = set(df_orders['customer_id'].unique())
    init = len(df_customers)
    df_customers = df_customers[df_customers['customer_id'].isin(kept_customer_ids)].copy()
    print(f"Trimmed customers: {init - len(df_customers)} removed.")

# sellers y products que aparecen en items remanentes
if 'seller_id' in df_items.columns and 'df_sellers' in globals():
    kept_seller_ids = set(df_items['seller_id'].unique())
    init = len(df_sellers)
    df_sellers = df_sellers[df_sellers['seller_id'].isin(kept_seller_ids)].copy()
    print(f"Trimmed sellers: {init - len(df_sellers)} removed.")

if 'product_id' in df_items.columns and 'df_products' in globals():
    kept_product_ids = set(df_items['product_id'].unique())
    init = len(df_products)
    df_products = df_products[df_products['product_id'].isin(kept_product_ids)].copy()
    print(f"Trimmed products: {init - len(df_products)} removed.")

# --- 4) Checks de consistencia -------------------------------------------------
# Ning√∫n DF dependiente debe contener order_id que no est√©n en df_orders
for name, d in [("df_items", df_items), ("df_reviews", df_reviews), ("df_payments", df_payments)]:
    assert set(d['order_id']).issubset(remaining_order_ids), f"{name} qued√≥ con order_id hu√©rfanos."

print("\nRemaining records:")
print(f"df_orders:   {len(df_orders)}")
print(f"df_items:    {len(df_items)}")
print(f"df_reviews:  {len(df_reviews)}")
print(f"df_payments: {len(df_payments)}")
```

| order\_status   | Condici√≥n para mantenerse                                                                                                                                                                               |
| --------------- | ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| **canceled**    | Se mantuvieron aunque tengan `approved_at`, `carrier_date` o `customer_date` nulos (porque es normal que no tengan esas fechas si fueron canceladas).                                                   |
| **unavailable** | Igual que canceled: se mantuvieron aunque las fechas sean nulas (porque nunca se pudieron despachar).                                                                                                   |
| **created**     | Mantienes los que tienen fechas nulas (esperado, nunca se procesaron).                                                                                                                                  |
| **delivered**   | Mantienes **solo los que tienen las tres fechas completas** (`approved_at`, `carrier_date`, `customer_date`). Los delivered con alguna fecha nula se eliminaron.                                        |
| **approved**    | Mantienes los que tienen `approved_at` (los v√°lidos). Eliminaste los que ten√≠an `approved_at` nulo (no tiene sentido estar en "approved" sin fecha de aprobaci√≥n) y los que ten√≠an `carrier_date` nulo. |
| **invoiced**    | Mantienes aunque tengan `carrier_date` y `customer_date` nulos (porque es l√≥gico, todav√≠a no se despachan).                                                                                             |
| **processing**  | Igual que invoiced, mantienes aunque tengan nulos.                                                                                                                                                      |
| **shipped**     | Mantienes aunque `customer_date` sea nulo (porque es l√≥gico, est√°n en tr√°nsito).                                                                                                                        |


Sugerencia de centinelas (distintas por columna)

order_approved_at ‚Üí 2090-01-01

order_delivered_carrier_date ‚Üí 2090-01-02

order_delivered_customer_date ‚Üí 2090-01-03

```python
import pandas as pd
import numpy as np

# Asegura que las columnas sean datetime (por si vienen como string)
date_cols = ['order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date']
for c in date_cols:
    df_orders[c] = pd.to_datetime(df_orders[c], errors='coerce')

# Define centinelas
SENTINEL_APPROVED = pd.Timestamp('2090-01-01')
SENTINEL_CARRIER  = pd.Timestamp('2090-01-02')
SENTINEL_CUSTOMER = pd.Timestamp('2090-01-03')

# 1) approved_at nulo pero ESPERABLE (canceled, unavailable, created) -> centinela
mask_approved_expected_null = (
    df_orders['order_approved_at'].isna() &
    df_orders['order_status'].isin(['canceled', 'unavailable', 'created'])
)
df_orders['approved_was_null'] = False
df_orders.loc[mask_approved_expected_null, 'approved_was_null'] = True
df_orders.loc[mask_approved_expected_null, 'order_approved_at'] = SENTINEL_APPROVED

# 2) carrier_date nulo pero ESPERABLE (canceled, unavailable, processing, invoiced) -> centinela
mask_carrier_expected_null = (
    df_orders['order_delivered_carrier_date'].isna() &
    df_orders['order_status'].isin(['canceled', 'unavailable', 'processing', 'invoiced'])
)
df_orders['carrier_was_null'] = False
df_orders.loc[mask_carrier_expected_null, 'carrier_was_null'] = True
df_orders.loc[mask_carrier_expected_null, 'order_delivered_carrier_date'] = SENTINEL_CARRIER

# 3) customer_date nulo pero ESPERABLE (canceled, unavailable, shipped) -> centinela
mask_customer_expected_null = (
    df_orders['order_delivered_customer_date'].isna() &
    df_orders['order_status'].isin(['canceled', 'unavailable', 'shipped'])
)
df_orders['customer_was_null'] = False
df_orders.loc[mask_customer_expected_null, 'customer_was_null'] = True
df_orders.loc[mask_customer_expected_null, 'order_delivered_customer_date'] = SENTINEL_CUSTOMER


# (Opcional) columnas helper para detectar centinelas de un vistazo
df_orders['is_sentinel_approved'] = df_orders['order_approved_at'].eq(SENTINEL_APPROVED)
df_orders['is_sentinel_carrier']  = df_orders['order_delivered_carrier_date'].eq(SENTINEL_CARRIER)
df_orders['is_sentinel_customer'] = df_orders['order_delivered_customer_date'].eq(SENTINEL_CUSTOMER)
```

```python
# Centinela ya definido antes
SENTINEL_CUSTOMER = pd.Timestamp('2090-01-03')

# Ampliar la m√°scara de nulo esperable en customer_date
mask_customer_expected_null = (
    df_orders['order_delivered_customer_date'].isna() &
    df_orders['order_status'].isin([
        'canceled', 'unavailable', 'shipped', 'invoiced', 'processing', 'approved', 'created'
    ])
)

# Si no existen las banderas, cr√©alas (idempotente)
if 'customer_was_null' not in df_orders.columns:
    df_orders['customer_was_null'] = False
if 'is_sentinel_customer' not in df_orders.columns:
    df_orders['is_sentinel_customer'] = False

# Marcar y rellenar con centinela
df_orders.loc[mask_customer_expected_null, 'customer_was_null'] = True
df_orders.loc[mask_customer_expected_null, 'order_delivered_customer_date'] = SENTINEL_CUSTOMER
df_orders.loc[mask_customer_expected_null, 'is_sentinel_customer'] = True
```

```python
import pandas as pd
import numpy as np

# Diccionario para mapear nombres
dfs_pre = {
    "orders": df_orders_pre,
    "items": df_items_pre,
    "reviews": df_reviews_pre,
    "payments": df_payments_pre,
    "customers": df_customers_pre,
    "products": df_products_pre,
    "sellers": df_sellers_pre,
    "category": df_category_pre,
    "geolocation": df_geolocation_pre
}

dfs_post = {
    "orders": df_orders,
    "items": df_items,
    "reviews": df_reviews,
    "payments": df_payments,
    "customers": df_customers,
    "products": df_products,
    "sellers": df_sellers,
    "category": df_category,
    "geolocation": df_geolocation
}

summary_rows = []

for name, df_pre in dfs_pre.items():
    df_post = dfs_post[name]
    row = {
        "DataFrame": name,
        "Registros antes": len(df_pre),
        "Registros despu√©s": len(df_post),
        "Variaci√≥n (%)": (len(df_post) - len(df_pre)) / len(df_pre) * 100,
        "Columnas": len(df_post.columns)
    }

    # Conteo total de nulos antes / despu√©s
    row["Nulos antes"] = int(df_pre.isnull().sum().sum())
    row["Nulos despu√©s"] = int(df_post.isnull().sum().sum())

    # IDs √∫nicos relevantes
    if "order_id" in df_pre.columns:
        row["√ìrdenes √∫nicas antes"] = df_pre["order_id"].nunique()
        row["√ìrdenes √∫nicas despu√©s"] = df_post["order_id"].nunique()
    if "customer_id" in df_pre.columns:
        row["Clientes √∫nicos antes"] = df_pre["customer_id"].nunique()
        row["Clientes √∫nicos despu√©s"] = df_post["customer_id"].nunique()

    # Promedios (solo en algunos DF)
    if name == "items":
        row["Promedio items/orden antes"] = df_pre.groupby("order_id").size().mean()
        row["Promedio items/orden despu√©s"] = df_post.groupby("order_id").size().mean()
    if name == "payments":
        row["Promedio pagos/orden antes"] = df_pre.groupby("order_id").size().mean()
        row["Promedio pagos/orden despu√©s"] = df_post.groupby("order_id").size().mean()
    if name == "reviews":
        row["Promedio reviews/orden antes"] = df_pre.groupby("order_id").size().mean()
        row["Promedio reviews/orden despu√©s"] = df_post.groupby("order_id").size().mean()


    summary_rows.append(row)

df_summary = pd.DataFrame(summary_rows)

# Ordenar columnas para legibilidad
cols_order = [
    "DataFrame", "Registros antes", "Registros despu√©s", "Variaci√≥n (%)",
    "Columnas", "Nulos antes", "Nulos despu√©s",
    "√ìrdenes √∫nicas antes", "√ìrdenes √∫nicas despu√©s",
    "Clientes √∫nicos antes", "Clientes √∫nicos despu√©s",
    "Promedio items/orden antes", "Promedio items/orden despu√©s",
    "Promedio pagos/orden antes", "Promedio pagos/orden despu√©s",
    "Promedio reviews/orden antes", "Promedio reviews/orden despu√©s",
]
df_summary = df_summary.reindex(columns=[c for c in cols_order if c in df_summary.columns])

# Display the DataFrame using the standard display function
display(df_summary)
```
